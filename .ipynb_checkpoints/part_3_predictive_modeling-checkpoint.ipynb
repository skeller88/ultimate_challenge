# Data cleaning and exploration

Data cleaning consisted of creating the "is_retained" target variable, dropping duplicate rows, imputing missing values with the median value, decomposing the date variables into year, month, and day, and 1 hot encoding the categorical variables. Data exploration consisted of visualizing the distributions of the variables, and the correlation of each predictor variable with the target variable.

```python
# Create "is_retained" target variable. Assume data was pulled on the most recent last_trip_date.
last_trip_date_dt = pd.to_datetime(users['last_trip_date'])
diff = (last_trip_date_dt.max() - last_trip_date_dt).dt.days
is_retained = diff <= 30
is_retained_true = is_retained[is_retained]
print(len(is_retained_true), len(is_retained), '% of users retained', len(is_retained_true) / len(is_retained) * 100)
users['is_retained'] = is_retained

# Make sure the distribution makes sense
diff.hist()
```

```python
def check_missing_values(df):
    # Check % missing values
    for column in df.columns:
        missing = df[df[column].isna()]
        if len(missing) > 0:
            print(column, len(missing), len(missing) / len(df) * 100)
        
check_missing_values(users)

# Check for duplicate values and drop
joined = users.apply(lambda row: "-".join(row.astype(str)), axis=1)
dups = joined[joined.duplicated()]
print(dups)
users_deduped = users[~users.index.isin(dups.index)]
print('dropped', len(users) - len(users_deduped), 'duplicated rows')
print(len(users_deduped))

missing_phones = users_deduped[users_deduped['phone'].isna()]
users_deduped.loc[missing_phones.index, 'phone'] = 'Unknown'
```

```python
# 1 hot encode the categorial variables
print('cities', users_deduped['city'].unique())
print('phone types', users_deduped['phone'].unique())

dfs_to_concat = [users_deduped]

for column in ['city', 'phone']:
    vals = users_deduped[column].values.reshape(-1, 1)
    encoder = OneHotEncoder().fit(vals)
    encoded = encoder.transform(vals).toarray()
    dfs_to_concat.append(pd.DataFrame(encoded, index=users_deduped.index, columns=encoder.categories_[0]))

users_with_dummies = pd.concat(dfs_to_concat, axis=1)

# Decompose date variables
for date_column in ['signup_date', 'last_trip_date']:
    datetime_column = pd.to_datetime(users_with_dummies[date_column]).dt
    for date_part in ['year', 'month', 'day']:
        users_with_dummies[f"{date_column}_{date_part}"] = getattr(datetime_column, date_part)

# Impute missing variables. Assume MCAR
def impute_missing(df, missing_columns):
    for column in missing_columns:)
        print(column, 'median', df[column].median())
        missing = df[df[column].isna()]
        df.loc[missing.index, column] = df[column].median()
    return df


users_with_dummies = impute_missing(users_with_dummies, ['avg_rating_of_driver', 'avg_rating_by_driver'])        
users_with_dummies = users_with_dummies.drop(['city', 'phone', 'signup_date', 'last_trip_date'], axis=1)

print('\n')
check_missing_values(users_with_dummies)
```
